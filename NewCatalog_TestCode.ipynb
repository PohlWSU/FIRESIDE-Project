```python
def line_lum_conv(line, fit_data):

    lir_lum = np.ma.masked_array(line['LIR'], mask=(line['LIR']==0))*u.solLum.to('erg s^-1')
    log_lir = np.ma.log10(lir_lum/(1e41*u.erg/u.s))
    log_line_lum= fit_data['coef']*log_lir+fit_data['y-int']    
    line_lum = np.ma.masked_array(((10**(log_line_lum)*1e41*u.erg/u.s).to('solLum')).value, mask = lir_lum.mask)

    return line_lum
    
results_table = []
for lines in spinoglio_lines:
    results = line_lum_conv(evt_data, lines) 
    results_table.append(results)
  

print(results)

all_elements= Table(results_table, names = ['[CII] 158',
                                            '[OI] 145',
                                            '[NII] 122',
                                            '[OIII] 88',
                                            '[OI] 63', 
                                            '[NIII] 57',
                                            '[OIII] 52',
                                            '[SIII] 19' ,
                                            '[NeIII] 16', 
                                            '[NeII] 13', 
                                            '[OIV] 26' , 
                                            'S[III] 33', 
                                            'S[II] 35'])
all_elements
```
