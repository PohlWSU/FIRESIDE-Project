```python
import pandas as pd
import plotly.express as px
import numpy as np
from dash import Dash, html, dash_table, dcc, callback, Output, Input


#Load data
df = np.load('/data/projects/FIRESIDE/data/line_table_cat_v9.npy', allow_pickle=True)
pio.renderers.default="iframe"

#Intialize app
app = Dash('test')
column_map = {"C[II] 158":df[11, 1:100], "O[I] 145":df[12, 1:100]}
#App layout
app.layout = html.Div([
    html.Div(children='C[II] 158 & O[I] 145 Data'),
    html.Hr(),
    dcc.RadioItems(options=({"label":k, "value":v} for k, v, in column_map.items()), value=df[11, 1:100], id='buttons'), dcc.Graph(figure={}, id='controls-and-graph')
])



# Add controls to build the interaction
@callback(
    Output(component_id='controls-and-graph', component_property='figure'),
    Input(component_id='buttons', component_property='value')
)
def update_graph(col_chosen):
    col_data = df[:, col_chosen, 1:100]
    fig = px.histogram(x=col_data, nbins=30, labels={'x':f'{col_chosen}'})
    fig.update_layout(title=f'{col_chosen}')
    return fig
    

    # Run the app
if __name__ == '__main__':
    app.run(debug=True, port=8051, mode='inline')


```
